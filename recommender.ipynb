## Popularity based recommender system

```python
# merging ratings and books using ISBN
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
# grouping books by Book-Title
number_ratings_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
number_ratings_df.rename(columns={"Book-Rating": "num_ratings"}, inplace=True)
number_ratings_df
```

```python
avg_number_ratings_df = ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_number_ratings_df.rename(columns={"Book-Rating": "avg_num_ratings"}, inplace=True)
avg_number_ratings_df
```

```python
# merge both dataframes avg_number_ratings_df and number_ratings_df
merged_df = number_ratings_df.merge(avg_number_ratings_df, on="Book-Title")
merged_df
```

```python
# get only those books with more than 250 ratings
popular_books = merged_df[merged_df["num_ratings"] > 250]
popular_books
```

```python
# sorting the dataframe by avg_num_ratings
popular_books = popular_books.sort_values("avg_num_ratings", ascending=False).head(50)

# get top 50 books
popular_books
```

```python
# get books authors, publishers and year
popular_books_info = popular_books.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-M", "num_ratings", "avg_num_ratings"]]
popular_books_info
```

## Collaborative filtering based recommender system 

```python
# get users ratings who have rated more than 200 books
users_with_ratings = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200
users_with_ratings_data = users_with_ratings[users_with_ratings].index
```

```python
# filter ratings dataframe to get only users who have rated more than 200 books
users_filtered_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(users_with_ratings_data)]
users_filtered_ratings
```

```python
# get each books by user rating and group by book title that has more than 50 ratings
book_title_filtered_ratings = users_filtered_ratings.groupby("Book-Title").count()["Book-Rating"] >= 50
book_title_filtered_ratings = book_title_filtered_ratings[book_title_filtered_ratings].index
book_title_filtered_ratings
```

```python
final_filtered_ratings = users_filtered_ratings[users_filtered_ratings["Book-Title"].isin(book_title_filtered_ratings)]
final_filtered_ratings
```

```python
# getting data pivot table
p_table = final_filtered_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating").fillna(0)
```

```python
# ratings of each user for each book
p_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# gettting cosine similarity matrix by calculating the euclidean distance between each user's ratings
similarity_matrix = cosine_similarity(p_table)
similarity_matrix, similarity_matrix.shape
```

```python
# viewing the first book's similarity with other books
similarity_matrix[0]
```

```python
# recommender function
def get_recommendations(book_name):
    # get index of the book
    book_index = np.where(p_table.index == book_name)[0][0]
    # get similar books
    similar_books = sorted(enumerate(similarity_matrix[book_index]), key=lambda x: x[1], reverse=True)[1:6]
    # get recommended books
    data = []
    for i in similar_books:
        item = []
        temp_item = books[books["Book-Title"] == p_table.index[i[0]]].drop_duplicates("Book-Title")
        item.extend(list(temp_item["Book-Title"].values))
        item.extend(list(temp_item["Book-Author"].values))
        item.extend(list(temp_item["Image-URL-M"].values))
        data.append(item)
    return data
```
